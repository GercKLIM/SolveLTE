```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = Image.open(os.path.join(folder, filename))
        if img is not None:
            images.append(img)
    return images

def show_image_gallery(images, cols=2):
    n_images = len(images)
    nrows = (n_images + cols - 1) // cols  # Calculate rows needed
    
    fig, axes = plt.subplots(nrows, cols, figsize=(15, nrows * 5))
    for i, ax in enumerate(axes.flat):
        if i < n_images:
            ax.imshow(images[i])
            ax.axis('off')
        else:
            ax.axis('off')

    plt.show()

# Путь к папке с изображениями
folder_path = "C:/Users/gerce/Documents/WORK DIRECTORY/Курсовая работа 6 сем/КОД/Создание датасетов/dataset_№1/train/test5"
images = load_images_from_folder(folder_path)

show_image_gallery(images, cols=3)
```
